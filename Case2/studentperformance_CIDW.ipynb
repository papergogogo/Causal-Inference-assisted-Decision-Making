```python
import os
import pickle
from tqdm import tqdm
import torch
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pyro
import pyro.distributions as dist
from torch import nn
import seaborn as sns
from pyro.nn import PyroModule
from sklearn.model_selection import train_test_split
```

```python
# for CI testing
# smoke_test = ('CI' in os.environ)
pyro.enable_validation(True)
pyro.set_rng_seed(1)
pyro.enable_validation(True)

# setup
assert issubclass(PyroModule[nn.Linear], nn.Linear)
assert issubclass(PyroModule[nn.Linear], PyroModule)
```

###Create model

```python
def StudentScoresModel(gender, race, ple,     # å¤–ç”Ÿå˜é‡ (scalars)
                       RS=None, MS=None, WS=None):    # å¯ä»¥ä»»é€‰åœ°è§‚æµ‹
    """
    æ–¹ç¨‹ï¼š
      MS ~ N(-11.767*G + 0.935*R , 1)
      RS ~ N( 7.009*G + 1.672*R + 1.131*MS - 1.123*PLE , 1)
      WS ~ N( 5.277*G + 0.259*MS + 0.724*RS - 0.308*PLE , 1)
    ç³»æ•°å›ºå®šï¼Œå™ªå£°æ–¹å·®å– 1ï¼Œå¦‚éœ€å­¦ä¹ å¯æ”¹æˆéšæœºã€‚
    """
    # â”€â”€ 1. math_score â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    MS_mean = -11.767 * gender + 0.935 * race
    MS = pyro.sample("MS", dist.Normal(MS_mean, 1.0), obs=MS)

    # â”€â”€ 2. reading_score â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    RS_mean = (7.009 * gender +
               1.672 * race   +
               1.131 * MS     -
               1.123 * ple)
    RS = pyro.sample("RS", dist.Normal(RS_mean, 1.0), obs=RS)

    # â”€â”€ 3. writing_score â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    WS_mean = (5.277 * gender +
               0.259 * MS     +
               0.724 * RS     -
               0.308 * ple)
    WS = pyro.sample("WS", dist.Normal(WS_mean, 1.0), obs=WS)

    return MS, WS            # æ–¹ä¾¿ direct call æ—¶å–å€¼
```

```python
def infer_single(row, num_samples=1_000):
    """
    ç»™å®šä¸€è¡Œ DataFrameï¼Œåªè§‚æµ‹ reading_scoreï¼Œ
    è¿”å›ž math_scoreã€writing_score çš„åŽéªŒæ ·æœ¬å‡å€¼
    """
    # scalar tensors
    G   = torch.tensor(row["gender"],  dtype=torch.float32)
    R   = torch.tensor(row["race"],    dtype=torch.float32)
    PLE = torch.tensor(row["parental_level_of_education"],
                       dtype=torch.float32)
    RS_obs = torch.tensor(row["reading_score"], dtype=torch.float32)

    # ç”¨ condition å›ºå®š RS
    conditioned_model = pyro.condition(
        StudentScoresModel,
        data={"RS": RS_obs}
    )

    # Importance é‡‡æ · â€”â€” ä¸Žæ‚¨åŽŸæ¥çš„å†™æ³•ä¸€è‡´
    imp = pyro.infer.Importance(conditioned_model,
                                num_samples=num_samples)

    posterior_traces = imp.run(gender=G, race=R, ple=PLE)

    # æŠ½å–åŽéªŒ
    MS_marg = pyro.infer.EmpiricalMarginal(posterior_traces, sites="MS")
    WS_marg = pyro.infer.EmpiricalMarginal(posterior_traces, sites="WS")

    return MS_marg.mean.item(), WS_marg.mean.item()
```

```python
math_pred, writing_pred = [], []

for _, row in tqdm(df.iterrows(), total=len(df)):
    m, w = infer_single(row, num_samples=1_000)
    math_pred.append(m)
    writing_pred.append(w)

df["math_score"]    = math_pred
df["writing_score"] = writing_pred
```

```python
from pyro.infer import Predictive

def infer_batch(dataframe, num_samples=1_000):
    G   = torch.tensor(dataframe["gender"].values,  dtype=torch.float32)
    R   = torch.tensor(dataframe["race"].values,    dtype=torch.float32)
    PLE = torch.tensor(dataframe["parental_level_of_education"].values,
                       dtype=torch.float32)
    RS_obs = torch.tensor(dataframe["reading_score"].values,
                          dtype=torch.float32)

    # æ¡ä»¶åŒ–ï¼šä»…ç»™å®š RS
    conditioned = pyro.condition(
        StudentScoresModel,
        data={"RS": RS_obs}
    )

    predictive = Predictive(conditioned,
                            num_samples=num_samples,
                            return_sites=("MS", "WS"))

    samples = predictive(gender=G, race=R, ple=PLE)  # dict é‡Œå« (num_samples, N)

    # å–å‡å€¼ä½œä¸ºç‚¹ä¼°è®¡
    math_mean    = samples["MS"].mean(0).numpy()
    writing_mean = samples["WS"].mean(0).numpy()

    dataframe = dataframe.copy()
    dataframe["math_imputed"]    = math_mean
    dataframe["writing_imputed"] = writing_mean
    return dataframe, samples    # å¦‚éœ€å…¨å¥—æ ·æœ¬å¯åŒæ—¶è¿”å›ž
```
