```python
import os
import pickle
from tqdm import tqdm
import torch
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pyro
import pyro.distributions as dist
from torch import nn
import seaborn as sns
from pyro.nn import PyroModule
from sklearn.model_selection import train_test_split
```

###Create model

```python
def ScoresModelMSWS(G, R, PLE,           # 外生变量 (tensor 同长度)
                    RS=None, MS=None, WS=None):
    """
    与图匹配的结构方程 (所有 λ 都视为需要学习的权重)：

       MS =  λ1·G + λ2·R + λ3·PLE + ε_MS
       RS =  λ4·G + λ5·R + λ6·PLE + λ7·MS + ε_RS
       WS =  λ8·G + λ9·PLE + λ10·MS + λ11·RS + ε_WS
    噪声方差固定为 1，也可以再放 HalfNormal/InverseGamma 先验。
    """

    stdn = dist.Normal(0., 1.)

    # --- 所有权重 (λ) 与截距 (b) ----
    b_MS      = pyro.sample("b_MS",      stdn)
    lam_G_MS  = pyro.sample("lam_G_MS",  stdn)
    lam_R_MS  = pyro.sample("lam_R_MS",  stdn)
    lam_PLE_MS= pyro.sample("lam_PLE_MS",stdn)

    b_RS      = pyro.sample("b_RS",      stdn)
    lam_G_RS  = pyro.sample("lam_G_RS",  stdn)
    lam_R_RS  = pyro.sample("lam_R_RS",  stdn)
    lam_PLE_RS= pyro.sample("lam_PLE_RS",stdn)
    lam_MS_RS = pyro.sample("lam_MS_RS", stdn)

    b_WS      = pyro.sample("b_WS",      stdn)
    lam_G_WS  = pyro.sample("lam_G_WS",  stdn)
    lam_PLE_WS= pyro.sample("lam_PLE_WS",stdn)
    lam_MS_WS = pyro.sample("lam_MS_WS", stdn)
    lam_RS_WS = pyro.sample("lam_RS_WS", stdn)

    # ------------- 逐学生并行 -------------
    N = len(G)
    with pyro.plate("students", N):

        # ---- MS ----
        MS_mean = (b_MS +
                   lam_G_MS   * G +
                   lam_R_MS   * R +
                   lam_PLE_MS * PLE)
        MS = pyro.sample("MS", dist.Normal(MS_mean, 1.), obs=MS)

        # ---- RS ----
        RS_mean = (b_RS +
                   lam_G_RS   * G +
                   lam_R_RS   * R +
                   lam_PLE_RS * PLE +
                   lam_MS_RS  * MS)
        RS = pyro.sample("RS", dist.Normal(RS_mean, 1.), obs=RS)

        # ---- WS ----
        WS_mean = (b_WS +
                   lam_G_WS   * G +
                   lam_PLE_WS * PLE +
                   lam_MS_WS  * MS +
                   lam_RS_WS  * RS)
        WS = pyro.sample("WS", dist.Normal(WS_mean, 1.), obs=WS)

    return MS, WS            # 方便直接取值
```

```python
from tqdm.auto import tqdm
import pickle

# ----------- DataFrame 准备 ---------------
# 确保列顺序/命名与下面一致

# 把类别变量提前数值化，例如 0/1；哑变量可以取其中一列
G_tensor   = torch.tensor(df["gender"].values,  dtype=torch.float32)
R_tensor   = torch.tensor(df["race"].values,    dtype=torch.float32)
PLE_tensor = torch.tensor(df["parental_level_of_education"].values,
                          dtype=torch.float32)
RS_tensor  = torch.tensor(df["reading_score"].values, dtype=torch.float32)

# 用来存放后验均值
MS_post_mean, WS_post_mean = [], []

for i in tqdm(range(len(df))):

    # condition 里只固定读写分数；外生变量直接按标量传
    conditioned = pyro.condition(
        ScoresModelMSWS,
        data={"RS": RS_tensor[i]}      # 只观测 reading_score
    )

    posterior = pyro.infer.Importance(conditioned,
                                      num_samples=1_000)

    trace_emp = posterior.run(
        G  = G_tensor[i:i+1],          # 形状保持一致
        R  = R_tensor[i:i+1],
        PLE= PLE_tensor[i:i+1]
    )

    MS_marg = pyro.infer.EmpiricalMarginal(trace_emp, sites="MS")
    WS_marg = pyro.infer.EmpiricalMarginal(trace_emp, sites="WS")

    MS_post_mean.append(MS_marg.mean.item())
    WS_post_mean.append(WS_marg.mean.item())

# 收回 DataFrame
df["math_imputed"]    = MS_post_mean
df["writing_imputed"] = WS_post_mean

# 如有需要保存
with open("inferred_scores.pkl", "wb") as f:
    pickle.dump({"math": MS_post_mean,
                 "writing": WS_post_mean}, f)
```

```python
from pyro.infer import Predictive

def infer_all(df, num_samples=2_000):
    G   = torch.tensor(df["gender"].values,  dtype=torch.float32)
    R   = torch.tensor(df["race"].values,    dtype=torch.float32)
    PLE = torch.tensor(df["parental_level_of_education"].values,
                       dtype=torch.float32)
    RS_obs = torch.tensor(df["reading_score"].values, dtype=torch.float32)

    conditioned = pyro.condition(ScoresModelMSWS, data={"RS": RS_obs})

    predictive = Predictive(conditioned,
                            num_samples=num_samples,
                            return_sites=("MS", "WS"))

    samples = predictive(G=G, R=R, PLE=PLE)   # (S, N)
    df["math_imputed"]    = samples["MS"].mean(0).numpy()
    df["writing_imputed"] = samples["WS"].mean(0).numpy()
    return df

df_pred = infer_all(df)
```
